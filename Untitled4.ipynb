```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Veri seti yolu (kagglehub ile indirilen)
src_dir = "/root/.cache/kagglehub/datasets/alessiocorrado99/animals10/versions/2/raw-img"
dest_dir = "./animals_4class2"

classes = {
    'dog': 'cane',
    'cat': 'gatto',
    'elephant': 'elefante',
    'chicken': 'gallina'
}

for split in ['train', 'validation']:
    os.makedirs(os.path.join(dest_dir, split), exist_ok=True)
    for eng_name in classes.keys():
        os.makedirs(os.path.join(dest_dir, split, eng_name), exist_ok=True)

for eng_name, it_name in classes.items():
    src_cls_dir = os.path.join(src_dir, it_name)

    if not os.path.exists(src_cls_dir):
        print(f"Hata: {it_name} klasörü bulunamadı!")
        continue

    files = [f for f in os.listdir(src_cls_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    # En fazla 500 örnekle sınırla (class balance için)
    files = files[:1000]
    train_files, val_files = train_test_split(files, test_size=0.2, random_state=42)

    # Kopyalama işlemi
    for file in train_files:
        shutil.copy(os.path.join(src_cls_dir, file),
                  os.path.join(dest_dir, 'train', eng_name, file))

    for file in val_files:
        shutil.copy(os.path.join(src_cls_dir, file),
                  os.path.join(dest_dir, 'validation', eng_name, file))

print("4 sınıflı veri seti hazır!")
print("Eğitim örnekleri:")
!ls {os.path.join(dest_dir, 'train')}/* | head -5
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
# Eğitim için Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Validasyon için sadece rescale
val_datagen = ImageDataGenerator(rescale=1./255)

# Generators
train_generator = train_datagen.flow_from_directory(
    './animals_4class2/train',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = val_datagen.flow_from_directory(
    './animals_4class2/validation',
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Sınıf isimlerini ve sayılarını kontrol
print(f"Sınıflar: {train_generator.class_indices}")
print(f"Toplam eğitim örnekleri: {train_generator.samples}")
print(f"Toplam validasyon örnekleri: {validation_generator.samples}")
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(4, activation='softmax')  # 4 sınıf
])

model.compile(optimizer=Adam(learning_rate=0.0001),
             loss='categorical_crossentropy',
             metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=35,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    verbose=1
)
```

```python
# Accuracy grafiği
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Metrics')
plt.legend()

# Loss grafiği
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Metrics')
plt.legend()
plt.show()
```

```python
model.save('animal_classifier_4class1.h5')
print("Model başarıyla kaydedildi!")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
from tensorflow.keras.models import load_model

# 1. Fine-tuned modeli yükle
model = load_model('fine_tuned_animal_mode22l.h5')  # Kaydettiğiniz model dosyası

# 2. Kullanıcıdan görsel yükleme
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 3. Görsel ön işleme
def preprocess_image(img_path):
    img = Image.open(img_path).convert('RGB')
    img = np.array(img)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    img = cv2.resize(img, (150, 150))
    img = img / 255.0  # Normalizasyon
    return np.expand_dims(img, axis=0)

# 4. Tahmin fonksiyonu
def predict_and_visualize(img_path):
    # Görseli hazırla
    img_array = preprocess_image(img_path)

    # Tahmin yap
    predictions = model.predict(img_array)[0]
    class_names = ['cat', 'dog', 'elephant', 'chicken']  # Eğitimle AYNI sırada!

    # Orijinal görseli oku (görselleştirme için)
    img_display = cv2.imread(img_path)
    img_display = cv2.cvtColor(img_display, cv2.COLOR_BGR2RGB)

    # Sonuçları hazırla
    predicted_class = class_names[np.argmax(predictions)]
    confidence = np.max(predictions)

    # Görsel üzerine yazı ekle
    cv2.putText(img_display,
               f"{predicted_class} ({confidence:.1%})",
               (10, 30),
               cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Görseli göster
    plt.figure(figsize=(8,8))
    plt.imshow(img_display)
    plt.axis('off')
    plt.show()

    # Detaylı sonuçlar
    print("\n🔍 Detaylı Tahmin Olasılıkları:")
    for name, prob in zip(class_names, predictions):
        print(f"{name:<8}: {prob:.4f}")

# 5. Testi çalıştır
predict_and_visualize(filename)
```
